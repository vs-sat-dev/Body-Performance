```python
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import (accuracy_score, confusion_matrix, 
                             classification_report, f1_score,
                             precision_score, recall_score, roc_auc_score)
from xgboost import XGBClassifier
from xgboost import plot_importance
import numpy as np
```

```python
def xgb_run(data):
    full_preds = np.zeros(len(data))
    kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
    for fold, (trn_ind, val_ind) in enumerate(kfold.split(data.drop('class', axis=1), data['class'])):
        x_train, y_train = data.loc[trn_ind].drop('class', axis=1), data.loc[trn_ind, 'class']
        x_val, y_val = data.loc[val_ind].drop('class', axis=1), data.loc[val_ind, 'class']

        model = XGBClassifier(silent=True, verbosity=0)
        model.fit(x_train, y_train)
        predictions = model.predict(x_val)
        full_preds[y_val.index] = predictions

        print(f'\nfold: {fold + 1}, accuracy: {accuracy_score(predictions, y_val)}')
        print(f'fold: {fold + 1}, f1-score-macro: {f1_score(y_val, predictions, average="macro")}')
        print(f'fold: {fold + 1}, f1-score-weighted: {f1_score(y_val, predictions, average="weighted")}')
        if fold == 4:
            plot_importance(model)
        
    return full_preds
```

```python
# I used copypaste with little changes from here https://github.com/optuna/optuna-examples/blob/main/xgboost/xgboost_simple.py
import optuna
import xgboost as xgb
import numpy as np

def objective(trial, data, metric_func):
    pred_vals = np.zeros(len(data))
    kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
    for fold, (trn_ind, val_ind) in enumerate(kfold.split(data.drop('class', axis=1), data['class'])):
        train_x, train_y = data.loc[trn_ind].drop('class', axis=1), data.loc[trn_ind, 'class']
        valid_x, valid_y = data.loc[val_ind].drop('class', axis=1), data.loc[val_ind, 'class']
        dtrain = xgb.DMatrix(train_x, label=train_y)
        dvalid = xgb.DMatrix(valid_x, label=valid_y)

        param = {
            "verbosity": 0,
            #"objective": "multi:softprob",
            "objective": "binary:logistic",
            #"eval_metric": "merror",
            #"num_class": len(data['class'].unique()),
            # use exact for small dataset.
            "tree_method": "exact",
            # defines booster, gblinear for linear functions.
            "booster": trial.suggest_categorical("booster", ["gbtree", "gblinear", "dart"]),
            # L2 regularization weight.
            "lambda": trial.suggest_float("lambda", 1e-8, 1.0, log=True),
            # L1 regularization weight.
            "alpha": trial.suggest_float("alpha", 1e-8, 1.0, log=True),
            # sampling ratio for training data.
            "subsample": trial.suggest_float("subsample", 0.2, 1.0),
            # sampling according to each tree.
            "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
        }

        if param["booster"] in ["gbtree", "dart"]:
            # maximum depth of the tree, signifies complexity of the tree.
            param["max_depth"] = trial.suggest_int("max_depth", 3, 9, step=2)
            # minimum child weight, larger the term more conservative the tree.
            param["min_child_weight"] = trial.suggest_int("min_child_weight", 2, 10)
            param["eta"] = trial.suggest_float("eta", 1e-8, 1.0, log=True)
            # defines how selective algorithm is.
            param["gamma"] = trial.suggest_float("gamma", 1e-8, 1.0, log=True)
            param["grow_policy"] = trial.suggest_categorical("grow_policy", ["depthwise", "lossguide"])

        if param["booster"] == "dart":
            param["sample_type"] = trial.suggest_categorical("sample_type", ["uniform", "weighted"])
            param["normalize_type"] = trial.suggest_categorical("normalize_type", ["tree", "forest"])
            param["rate_drop"] = trial.suggest_float("rate_drop", 1e-8, 1.0, log=True)
            param["skip_drop"] = trial.suggest_float("skip_drop", 1e-8, 1.0, log=True)

        bst = xgb.train(param, dtrain)
        preds = bst.predict(dvalid)
        pred_labels = np.rint(preds)
        pred_vals[valid_y.index] = pred_labels
    
    return metric_func(data["class"], pred_vals)

def optuna_with_metric(data, metric_func, models=None):
    accuracy_vals = np.zeros(len(data))

    pred_vals = np.zeros(len(data))
    if models is None:
        objective_with_df = lambda trials: objective(trials, data, metric_func)
        study = optuna.create_study(direction="maximize")
        study.optimize(objective_with_df, n_trials=200)

        params = study.best_trial.params
        params['objective'] = 'binary:logistic'
        models = []
        kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
        for fold, (trn_ind, val_ind) in enumerate(kfold.split(data.drop('class', axis=1), data['class'])):
            train_x, train_y = data.loc[trn_ind].drop('class', axis=1), data.loc[trn_ind, 'class']
            valid_x, valid_y = data.loc[val_ind].drop('class', axis=1), data.loc[val_ind, 'class']
            dtrain = xgb.DMatrix(train_x, label=train_y)
            dvalid = xgb.DMatrix(valid_x, label=valid_y)

            models.append(xgb.train(params, dtrain))
            preds = models[fold].predict(dvalid)
            pred_labels = np.rint(preds)
            pred_vals[valid_y.index] = pred_labels
        
        return pred_vals, models
    else:
        dmatrix = xgb.DMatrix(data.drop('class', axis=1))
        for model in models:
            pred_vals += model.predict(dmatrix)
        pred_vals /= len(models)
        pred_vals = np.rint(pred_vals)
        
        return pred_vals, models
```

```python
def objective2(trial, data, metric_func, avrg):
    pred_vals = np.zeros(len(data))
    kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
    for fold, (trn_ind, val_ind) in enumerate(kfold.split(data.drop('class', axis=1), data['class'])):
        train_x, train_y = data.loc[trn_ind].drop('class', axis=1), data.loc[trn_ind, 'class']
        valid_x, valid_y = data.loc[val_ind].drop('class', axis=1), data.loc[val_ind, 'class']
        dtrain = xgb.DMatrix(train_x, label=train_y)
        dvalid = xgb.DMatrix(valid_x, label=valid_y)

        param = {
            "verbosity": 0,
            "objective": "multi:softprob",
            #"objective": "binary:logistic",
            #"eval_metric": "merror",
            "num_class": len(data['class'].unique()),
            # use exact for small dataset.
            #"tree_method": "exact",
            # defines booster, gblinear for linear functions.
            "booster": trial.suggest_categorical("booster", ["gbtree", "gblinear", "dart"]),
            # L2 regularization weight.
            "lambda": trial.suggest_float("lambda", 1e-8, 1.0, log=True),
            # L1 regularization weight.
            "alpha": trial.suggest_float("alpha", 1e-8, 1.0, log=True),
            # sampling ratio for training data.
            "subsample": trial.suggest_float("subsample", 0.2, 1.0),
            # sampling according to each tree.
            "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
        }

        if param["booster"] in ["gbtree", "dart"]:
            # maximum depth of the tree, signifies complexity of the tree.
            param["max_depth"] = trial.suggest_int("max_depth", 3, 9, step=2)
            # minimum child weight, larger the term more conservative the tree.
            param["min_child_weight"] = trial.suggest_int("min_child_weight", 2, 10)
            param["eta"] = trial.suggest_float("eta", 1e-8, 1.0, log=True)
            # defines how selective algorithm is.
            param["gamma"] = trial.suggest_float("gamma", 1e-8, 1.0, log=True)
            param["grow_policy"] = trial.suggest_categorical("grow_policy", ["depthwise", "lossguide"])

        if param["booster"] == "dart":
            param["sample_type"] = trial.suggest_categorical("sample_type", ["uniform", "weighted"])
            param["normalize_type"] = trial.suggest_categorical("normalize_type", ["tree", "forest"])
            param["rate_drop"] = trial.suggest_float("rate_drop", 1e-8, 1.0, log=True)
            param["skip_drop"] = trial.suggest_float("skip_drop", 1e-8, 1.0, log=True)

        bst = xgb.train(param, dtrain)
        preds = bst.predict(dvalid)
        pred_labels = np.rint(preds)
        pred_vals[valid_y.index] = np.argmax(pred_labels, axis=1)
        
    if avrg:
        return metric_func(data["class"], pred_vals, average=avrg)
    else:
        return metric_func(data["class"], pred_vals)
    

def optuna_multiclass(data, metric_func, avrg, models=None):
    accuracy_vals = np.zeros(len(data))

    pred_vals = np.zeros((len(data), 4))
    if models is None:
        objective_with_df = lambda trials: objective2(trials, data, metric_func, avrg)
        study = optuna.create_study(direction="maximize")
        study.optimize(objective_with_df, n_trials=200)

        params = study.best_trial.params
        params['objective'] = "multi:softprob"
        #params['eval_metric'] = 'merror'
        params['num_class'] = 4
        models = []
        kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
        for fold, (trn_ind, val_ind) in enumerate(kfold.split(data.drop('class', axis=1), data['class'])):
            train_x, train_y = data.loc[trn_ind].drop('class', axis=1), data.loc[trn_ind, 'class']
            valid_x, valid_y = data.loc[val_ind].drop('class', axis=1), data.loc[val_ind, 'class']
            dtrain = xgb.DMatrix(train_x, label=train_y)
            dvalid = xgb.DMatrix(valid_x, label=valid_y)

            models.append(xgb.train(params, dtrain))
            preds = models[fold].predict(dvalid)
            pred_vals[valid_y.index] = preds
        return pred_vals, models
    else:
        dmatrix = xgb.DMatrix(data.drop('class', axis=1))
        for model in models:
            pred_vals += model.predict(dmatrix)
        pred_vals /= len(models)
        pred_vals = np.rint(pred_vals)
        
        return pred_vals, models
```

```python
def xgb_prob_run(data, models=None):
    full_preds = np.zeros((len(data), 4))
    if models is None:
        models = []
        kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
        for fold, (trn_ind, val_ind) in enumerate(kfold.split(data.drop('class', axis=1), data['class'])):
            x_train, y_train = data.loc[trn_ind].drop('class', axis=1), data.loc[trn_ind, 'class']
            x_val, y_val = data.loc[val_ind].drop('class', axis=1), data.loc[val_ind, 'class']

            models.append(XGBClassifier(silent=True, verbosity=0))
            models[fold].fit(x_train, y_train)
            predictions = models[fold].predict_proba(x_val)
            full_preds[y_val.index] = predictions
        
        return full_preds, models
    else:
        for model in models:
            full_preds += model.predict_proba(data.drop('class', axis=1))
        full_preds /= len(models)
        full_preds = np.rint(full_preds)
        
        return full_preds, models
```

```python
def objective3(trial, data, abcd_targets, 
               all_vs_all_f1_preds, all_vs_all_accuracy_preds,
               all_vs_all_precision_preds, all_vs_all_recall_preds,
               all_vs_all_standard_preds):
    abcd_trial = trial.suggest_float("abcd_targets", 0.0, 1.0)
    f1_trial = trial.suggest_float("all_f1", 0.0, 1.0)
    accuracy_trial = trial.suggest_float("all_accuracy", 0.0, 1.0)
    precision_trial = trial.suggest_float("all_precision", 0.0, 1.0)
    recall_trial = trial.suggest_float("all_recall", 0.0, 1.0)
    standard_trial = trial.suggest_float("all_standard", 0.0, 1.0)
    
    all_weights = abcd_trial + f1_trial + accuracy_trial \
                  + precision_trial + recall_trial + standard_trial
    if all_weights < 1e-4:
        all_weights = 1e-4
    
    abcd_weight = abcd_trial / all_weights
    f1_weight = f1_trial / all_weights
    accuracy_weight = accuracy_trial / all_weights
    precision_weight = precision_trial / all_weights
    recall_weight = recall_trial / all_weights
    standard_weight = standard_trial / all_weights
    
    targets = (abcd_targets * abcd_weight) + (all_vs_all_f1_preds * f1_weight) \
              + (all_vs_all_accuracy_preds * accuracy_weight) \
              + (all_vs_all_precision_preds * precision_weight) \
              + (all_vs_all_recall_preds * recall_weight) \
              + (all_vs_all_standard_preds * standard_weight)
    target_preds = np.argmax(targets, axis=1)
    
    return accuracy_score(data["class"], target_preds)

def optuna_weights(data, abcd_targets, 
                   all_vs_all_f1_preds, all_vs_all_accuracy_preds,
                   all_vs_all_precision_preds, all_vs_all_recall_preds,
                   all_vs_all_standard_preds):
    objective_with_df = lambda trials: objective3(trials, data, abcd_targets,
                                                  all_vs_all_f1_preds, 
                                                  all_vs_all_accuracy_preds,
                                                  all_vs_all_precision_preds, 
                                                  all_vs_all_recall_preds,
                                                  all_vs_all_standard_preds)
    study = optuna.create_study(direction="maximize")
    study.optimize(objective_with_df, n_trials=1000)
    
    p = study.best_trial.params
    all_weights = p['abcd_targets'] + p['all_f1'] + p['all_accuracy'] \
                      + p['all_precision'] + p['all_recall'] + p['all_standard']
    if all_weights < 1e-4:
        all_weights = 1e-4

    abcd_weight = p['abcd_targets'] / all_weights
    f1_weight = p['all_f1'] / all_weights
    accuracy_weight = p['all_accuracy'] / all_weights
    precision_weight = p['all_precision'] / all_weights
    recall_weight = p['all_recall'] / all_weights
    standard_weight = p['all_standard'] / all_weights

    targets = (abcd_targets * abcd_weight) + (all_vs_all_f1_preds * f1_weight) \
              + (all_vs_all_accuracy_preds * accuracy_weight) \
              + (all_vs_all_precision_preds * precision_weight) \
              + (all_vs_all_recall_preds * recall_weight) \
              + (all_vs_all_standard_preds * standard_weight)
    target_preds = np.argmax(targets, axis=1)
    
    return target_preds, p
```

```python
def pipeline(data, all_models=None):
    a_vs_all_df = data.copy()
    a_vs_all_df['class'] = a_vs_all_df['class'].replace({0: 1, 1: 0, 2: 0, 3: 0})
    
    b_vs_all_df = data.copy()
    b_vs_all_df['class'] = b_vs_all_df['class'].replace({0: 0, 1: 1, 2: 0, 3: 0})
    
    c_vs_all_df = data.copy()
    c_vs_all_df['class'] = c_vs_all_df['class'].replace({0: 0, 1: 0, 2: 1, 3: 0})
    
    d_vs_all_df = data.copy()
    d_vs_all_df['class'] = d_vs_all_df['class'].replace({0: 0, 1: 0, 2: 0, 3: 1})
    
    checker = True
    if all_models is None:
        checker = False
        all_models = dict()
    
    if not checker:
        a_vs_all_models = None
        b_vs_all_models = None
        c_vs_all_models = None
        d_vs_all_models = None
    else:
        a_vs_all_models = all_models['a_vs_all_models']
        b_vs_all_models = all_models['b_vs_all_models']
        c_vs_all_models = all_models['c_vs_all_models']
        d_vs_all_models = all_models['d_vs_all_models']
    
    a_vs_all_preds, all_models['a_vs_all_models'] = optuna_with_metric(a_vs_all_df, 
                                                                       recall_score, 
                                                                       a_vs_all_models)
    b_vs_all_preds, all_models['b_vs_all_models'] = optuna_with_metric(b_vs_all_df, 
                                                                       f1_score, 
                                                                       b_vs_all_models)
    c_vs_all_preds, all_models['c_vs_all_models'] = optuna_with_metric(c_vs_all_df, 
                                                                       f1_score, 
                                                                       c_vs_all_models)
    d_vs_all_preds, all_models['d_vs_all_models'] = optuna_with_metric(d_vs_all_df, 
                                                                       f1_score, 
                                                                       d_vs_all_models)
    
    a_vs_all_preds
    a_vs_all_preds = np.where(a_vs_all_preds == 0.0, -1, a_vs_all_preds)
    
    b_vs_all_preds
    b_vs_all_preds = np.where(b_vs_all_preds == 0.0, -1, b_vs_all_preds)
    
    c_vs_all_preds
    c_vs_all_preds = np.where(c_vs_all_preds == 0.0, -1, c_vs_all_preds)
    
    d_vs_all_preds
    d_vs_all_preds = np.where(d_vs_all_preds == 0.0, -1, d_vs_all_preds)
    
    if not checker:
        all_vs_all_f1_models = None
        all_vs_all_accuracy_models = None
        all_vs_all_precision_models = None
        all_vs_all_recall_models = None
    else:
        all_vs_all_f1_models = all_models['all_vs_all_f1_models']
        all_vs_all_accuracy_models = all_models['all_vs_all_accuracy_models']
        all_vs_all_precision_models = all_models['all_vs_all_precision_models']
        all_vs_all_recall_models = all_models['all_vs_all_recall_models']
    
    all_vs_all_f1_preds, all_models['all_vs_all_f1_models'] = \
                                            optuna_multiclass(data, f1_score, "weighted",
                                                              all_vs_all_f1_models)
    all_vs_all_accuracy_preds, all_models['all_vs_all_accuracy_models'] = \
                                            optuna_multiclass(data, accuracy_score, None,
                                                              all_vs_all_accuracy_models)
    all_vs_all_recall_preds, all_models['all_vs_all_recall_models'] = \
                                            optuna_multiclass(data, recall_score, "weighted",
                                                              all_vs_all_recall_models)
    all_vs_all_precision_preds, all_models['all_vs_all_precision_models'] = \
                                            optuna_multiclass(data, precision_score, "weighted",
                                                              all_vs_all_precision_models)
    
    if not checker:
        all_vs_all_standard_models = None
    else:
        all_vs_all_standard_models = all_models['all_vs_all_standard_models']
    all_vs_all_standard_preds, all_models['all_vs_all_standard_models'] = \
                                            xgb_prob_run(data, all_vs_all_standard_models)
    
    a = np.expand_dims(a_vs_all_preds, axis=0)
    b = np.expand_dims(b_vs_all_preds, axis=0)
    c = np.expand_dims(c_vs_all_preds, axis=0)
    d = np.expand_dims(d_vs_all_preds, axis=0)
    abcd_targets = np.concatenate([a, b, c, d], axis=0).T
    
    p = None
    if not checker:
        target_preds, all_models['weights'] = optuna_weights(data, abcd_targets, 
                                                             all_vs_all_f1_preds, 
                                                             all_vs_all_accuracy_preds,
                                                             all_vs_all_precision_preds, 
                                                             all_vs_all_recall_preds,
                                                             all_vs_all_standard_preds)
    else:
        p = all_models['weights']
        all_weights = p['abcd_targets'] + p['all_f1'] + p['all_accuracy'] \
                    + p['all_precision'] + p['all_recall'] + p['all_standard']
        if all_weights < 1e-4:
            all_weights = 1e-4

        abcd_weight = p['abcd_targets'] / all_weights
        f1_weight = p['all_f1'] / all_weights
        accuracy_weight = p['all_accuracy'] / all_weights
        precision_weight = p['all_precision'] / all_weights
        recall_weight = p['all_recall'] / all_weights
        standard_weight = p['all_standard'] / all_weights

        a = np.expand_dims(a_vs_all_preds, axis=0)
        b = np.expand_dims(b_vs_all_preds, axis=0)
        c = np.expand_dims(c_vs_all_preds, axis=0)
        d = np.expand_dims(d_vs_all_preds, axis=0)
        abcd_targets = np.concatenate([a, b, c, d], axis=0).T
        targets = (abcd_targets * abcd_weight) + (all_vs_all_f1_preds * f1_weight) \
                  + (all_vs_all_accuracy_preds * accuracy_weight) \
                  + (all_vs_all_precision_preds * precision_weight) \
                  + (all_vs_all_recall_preds * recall_weight) \
                  + (all_vs_all_standard_preds * standard_weight)
        target_preds = np.argmax(targets, axis=1)

    return target_preds, all_models
```
